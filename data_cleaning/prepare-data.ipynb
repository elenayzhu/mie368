# Imports

```python
import pandas as pd
import numpy as np
```

# Load dataframes
We need to load csv files as dataframes for all of the months and then combine them all into one dataframe 

```python
jan = pd.read_csv("2024_JAN.csv")
feb = pd.read_csv("2024_FEB.csv")
mar = pd.read_csv("2024_MAR.csv")
apr = pd.read_csv("2024_APRIL.csv")
may = pd.read_csv("2024_MAY.csv")
jun = pd.read_csv("2024_JUN.csv")
jul = pd.read_csv("2024_JUL.csv")
aug = pd.read_csv("2024_AUG.csv")
sep = pd.read_csv("2024_SEP.csv")
oct = pd.read_csv("2024_OCT.csv")
nov = pd.read_csv("2024_NOV.csv")
dec = pd.read_csv("2024_DEC.csv")

# Combine into one dataframa

df = pd.concat(
    [jan, feb, mar, apr, may, jun, jul, aug, sep, oct, nov, dec],
    ignore_index=True
)
df.head()
```

Create a new csv file for the full year data

```python
df.to_csv("2024_FULL_YEAR.csv", index=False)
```
